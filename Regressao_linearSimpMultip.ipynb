```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
dados=pd.read_csv('cars.csv')
dados
```

```python
dados.drop(['Unnamed: 0'], axis=1)
```

```python
x=dados.iloc[:,1].values #indep , variavel speed
y=dados.iloc[:,0].values #dependent
x
```

```python
#correlacao de x e y alta e forte
correlacao=np.corrcoef(x,y)
correlacao
```

```python
x=x.reshape(-1,1) #transf x enm matriz
modelo=LinearRegression()
modelo.fit(x,y)
```

```python
modelo.intercept_ #intercepcao
```

```python
modelo.coef_ #inclinacao
```

```python
plt.scatter(x,y) #grafico de dispersao e linha de ajuste
plt.plot(x,modelo.predict(x), color="red")
```

```python
calculo= (modelo.intercept_ + modelo.coef_ *22)#calculo do modelo da veloc q levou pra parar a 22pes
calculo
```

```python
modelo.predict([[22]]) #usando o predict para chegar no msm calculo acima
```

```python
#grafico de outliers
visualizador= ResidualsPlot(modelo)
visualizador.fit(x,y)
visualizador.poof()
```
